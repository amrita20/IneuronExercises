Problem Statement 7: 
 
Customers arrive at a rate of 72 per hour to my shop. What is the probability of 𝑘 customers arriving in 4 minutes? a) 5 customers, b) not more than 3 customers, c) more than 3 customers. Give a pictorial representation of the same to validate your answer. 

```python
Let m denote the customr arriving rate in 4 minutes.  
m = ((72*4)/60) 

The probability of 𝑘 customers arriving in 4 minutes is given by:
P = (e^(-m)*m^k)/k! 
```

```python
import math 
m = 4.8 
#5 customers i.e, P(X=5)
k = 5
P5 = (math.exp(-m)*pow(m,k))/math.factorial(k)
print(P5) 
```

```python
# not more than 3 customers i.e, P(0=<X<=3)
P0_3 = ((math.exp(-m)*pow(m,0))/math.factorial(0))+((math.exp(-m)*pow(m,1))/math.factorial(1))+((math.exp(-m)*pow(m,2))/math.factorial(2))+((math.exp(-m)*pow(m,3))/math.factorial(3))
print(P0_3)
```

```python
# OR
P0_3 = 0 
for k in range(0,4):
    P0_3 += (math.exp(-m)*pow(m,k))/math.factorial(k)
print(P0_3)    
```

```python
# more than 3 customers i.e, P(3<X<=k) = 1 - P0_3
P = 1 - (((math.exp(-m)*pow(m,0))/math.factorial(0))+((math.exp(-m)*pow(m,1))/math.factorial(1))+((math.exp(-m)*pow(m,2))/math.factorial(2))+((math.exp(-m)*pow(m,3))/math.factorial(3)))
print(P)
```

```python
# OR
P0_3 = 0 
for k in range(0,4):
    P0_3 += (math.exp(-m)*pow(m,k))/math.factorial(k)
P = 1 - P0_3    
print(P)  
```
